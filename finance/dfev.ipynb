```python
import pandas as pd
```

```python
def format_dfev(df):
    df = df.loc[df['isin'].str.startswith('BR', na=False)]

    # cols
    df['ticker'] = df['ticker'].str.split().str[0]
    df['weight%'] = df['weight'] * 100
    df['weight_br%'] = df['principal'] / df['principal'].sum()
    df['weight_br%'] = (df['weight_br%'] * 100).round(2)
    df['shares'] = (df['shares'] / 100).fillna(0).astype(int)
    df['principal'] = df['principal'].fillna(0).astype(int)
    df = df.drop(['etf_ticker','coupon_rate', 'maturity_date', 'sedol', 'identifier', 'weight'], axis=1)
    
    return df.sort_values(by='market_value', ascending=False) \
        .reset_index(drop=True)
```

```python
# position change
share_ratio_col = 'share_ratio'
df_merged = pd.merge(df_new, df_old, on="ticker", suffixes=('_new', '_old'))
df_merged[share_ratio_col] = (df_merged.shares_new / df_merged.shares_old).round(2)
df_merged[df_merged[share_ratio_col] != 1].sort_values(by=share_ratio_col, ascending=False)
```

```python
# new stocks
df_new_only = pd.merge(df_new, df_old, on="ticker", how='left', suffixes=('_new', '_old'), indicator=True)
df_new_only.loc[df_new_only['_merge'] == 'left_only'] \
    .sort_values(by='market_value_new', ascending=False)
```

```python
# dropped stocks
df_old_only = pd.merge(df_new, df_old, on="ticker", how='right', suffixes=('_new', '_old'), indicator=True)
df_old_only.loc[df_old_only['_merge'] == 'right_only'] \
    .sort_values(by='market_value_old', ascending=False)
```
